```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

file = pd.read_csv('provincess.csv')


#I created a new dataframe that has statistics for certain factors in the 2012 year.
#Please do not get confused by the fact that it says 2016; it is from 2012.


#I used data from 2016 to make predicions about the youth crime severity of a random year.
#Then I created a column to calculate the difference between the actual and predicted number.

r = smf.gls('Q("Youth Crime Severity Index 2016")~Q("Rates of households with violence 2016")',data=file).fit()
#r.summary()
d={'provinces':['Manitoba','Quebec','Ontario','PEI','Nunavut','Saskatchewan','BC','NB','Yukon','Alberta','NS','Newfoundland','NT'],'Rates of households with violence 2016':[323.7,217.8,131,144.3,2355.5,410.3,189.9,190.1,811,238.5,172.5,215.8,1767.1],
                      "Daily Smokers per Region 2016":[15.4,18.5,14.3,18.3,48.8,15.2,10.7,18.6,22.8,16.7,18.3,21,28.6],
                       "Rate of Children in lone-parent families 2016":[17.3,16.6,17.3,16.2,28.4,17.4,15.7,16.1,21.9,14.6,17.9,15.9,22.9]
   ,'Actual Youth Crime Severity 2012':[144.46,64.95,67.42,70.85,341.07,207.52,57.51,80.28,178.72,79.86,111.75,65.42,385.62]}
df=pd.DataFrame(data=d)
df['Predicted Youth Crime Severity 2012']=r.predict(df)
df[['Predicted Youth Crime Severity 2012','Actual Youth Crime Severity 2012']].astype('int64')
df['Difference']=abs(df['Predicted Youth Crime Severity 2012']-df['Actual Youth Crime Severity 2012'])
df
```

```python
# I then summed up the total errors as well as total statistics in the actual youth severity to see roughly how accurate my predictions were.

#I thought that this was a fair way of measuring accuracy since the higher a given number is, the more likely it is for a prediction
#to be more off.
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

file = pd.read_csv('provincess.csv')
#file.dropna(axis=1)
#file = file.drop(file.columns[[8,9,11,12]],axis=1)
#file.corr(method='pearson')

#file = file.drop(file.columns[[8,9,11,12]],axis=1)
#file.corr(method='pearson')

r = smf.gls('Q("Youth Crime Severity Index 2016")~Q("Rates of households with violence 2016")',data=file).fit()
#r.summary()

#I created a new dataframe that has statistics for certain factors in the 2012 year.
#Please do not get confused by the fact that it says 2016; it is from 2012.

d={'provinces':['Manitoba','Quebec','Ontario','PEI','Nunavut','Saskatchewan','BC','NB','Yukon','Alberta','NS','Newfoundland','NT'],'Rates of households with violence 2016':[323.7,217.8,131,144.3,2355.5,410.3,189.9,190.1,811,238.5,172.5,215.8,1767.1],
                      "Daily Smokers per Region 2016":[15.4,18.5,14.3,18.3,48.8,15.2,10.7,18.6,22.8,16.7,18.3,21,28.6],
                       "Rate of Children in lone-parent families 2016":[17.3,16.6,17.3,16.2,28.4,17.4,15.7,16.1,21.9,14.6,17.9,15.9,22.9]
   ,'actual':[144.46,64.95,67.42,70.85,341.07,207.52,57.51,80.28,178.72,79.86,111.75,65.42,385.62]}
df=pd.DataFrame(data=d)
df['predicted']=r.predict(df)
df[['predicted','actual']].astype('int64')
df['difference']=abs(df['predicted']-df['actual'])
q=df['difference'].sum()
w=df['actual'].sum()
print(100-(q/w)*100)


#Predicting Youth Crime Severity Index of a random year gives us roughly a 76% accuracy rate.
```
